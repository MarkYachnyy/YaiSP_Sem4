<h2>ТЕОРИЯ</h2>
Движение тела, брошенного под углом к горизонту, — движение тела в двумерной системе координат (по двум осям) при изначальном направлении начальной скорости под углом к горизонту. Данное движение является сложным видом механического движения с криволинейной траекторией. Такие типы движений принято рассматривать в проекции на оси выбранной системы координат.

Классическая постановка задач на подобную тематику: тело бросили под углом α к горизонту с начальной скоростью v<span style="font-size:10px;">o</span>, найти различные параметры движения.

Рассмотрим само движение. После броска траектория движущегося тела представляет собой параболу (докажем позже). Выберем произвольную точку на параболе и укажем ускорение, с которым движется тело в данный момент (ускорение свободного падения). Направление данного ускорения — вертикально вниз. Проекции данного ускорения на ось ОХ (g<span style="font-size:10px;">x</span>=0), и на ось OY (g<span style="font-size:10px;">y</span> = 10 м/c^2).

Тогда, вдоль оси ОХ, тело движется равномерно (т.к. ускорение вдоль этой оси равно 0). Более сложным является движение тела вдоль оси OY: сначала и до высшей точки тело движется равнозамедленно, потом движется равноускоренно

1. Для рассмотрения движения тела вдоль осей, введём начальные скорости движения тела вдоль выбранных нами осей (рис. 3). На рисунке представлена часть траектории в самом начале движения. Начальные скорости движения вдоль осей обозначим v<span style="font-size:10px;">ox</span> и v<span style="font-size:10px;">oy</span>. Исходя из треугольника, катетами которого являются наши проекции (можно построить параллельным переносом), а гипотенузой — модуль вектора начальной скорости v<span style="font-size:10px;">o</span>, можем найти значения необходимых нам проекций:

v<span style="font-size:10px;">ox</span>=v<span style="font-size:10px;">o</span>*cos(α)

v<span style="font-size:10px;">oy</span>=v<span style="font-size:10px;">o</span>*sin(α)

2. Попробуем найти полное время полёта (t<span style="font-size:10px;">max</span>). Для этого воспользуемся тем, что вдоль оси OY тело движется равнозамедленно, а в верхней точке движение вдоль этой оси и вовсе останавливается. Таким образом, конечная скорость в этой точке вдоль оси OY равна 0. Тогда, исходя из движения:

0 = v<span style="font-size:10px;">oy</span> - g * t<span style="font-size:10px;">max</span>/2

t<span style="font-size:10px;">max</span>/2, так как время движения от начальной до высшей точки, и от высшей точки до конечной будет равно. Тогда:

t<span style="font-size:10px;">max</span> = 2v<span style="font-size:10px;">oy</span>/g

И учитывая выведенные формулы для проекций скоростей на оси:

t<span style="font-size:10px;">max</span> = 2v<span style="font-size:10px;">o</span>*sin(α)/g




3. Перейдём к вопросу о максимальной дальности броска в горизонтальном направлении (S<span style="font-size:10px;">max</span>). Вдоль горизонта тело движется равномерно (рис. 2). Тогда путь, проделанный телом за время t<span style="font-size:10px;">max</span>:

S<span style="font-size:10px;">max</span> = v<span style="font-size:10px;">ox</span>*t<span style="font-size:10px;">max</span>

Учитывая формулы для v<span style="font-size:10px;">ox</span> и t<span style="font-size:10px;">max</span>

S<span style="font-size:10px;">max</span> = v<span style="font-size:10px;">o</span>^2 * sin(2α)/g

4. Перейдём к максимальной высоте полёта (H<span style="font-size:10px;">max</span>). Данный параметр связан с движением тела вдоль оси OY, которое, как мы выяснили, является равноускоренным/равнозамедленным. Рассмотрим участок от наивысшей точки до конечной: для него вдоль соответствующей оси тело без начальной скорости движется с ускорением g в течение времени t<span style="font-size:10px;">max</span>/2, формируем уравнение:

H<span style="font-size:10px;">max</span> = g * (t<span style="font-size:10px;">max</span>/2)^2 / 2 = g * (t<span style="font-size:10px;">max</span>)^2 / 8

С учётом полученных ранее формул:

H<span style="font-size:10px;">max</span> = v<span style="font-size:10px;">o</span>^2 * sin^2(α) / 2g 

<strong>Таким образом, ряд параметров движения при броске под углом к горизонту можно вычислить, зная лишь начальные параметры броска.</strong>

<h2>РАСЧЁТ ДВИЖЕНИЯ</h2>

Теперь пришло время рассчитать движение под углом к горизонту на конкретных начальных параметрах, и посмотреть, будут ли совпадать рассчитанные по полученным формулам параметры движения с теми, что мы сможем увидеть из графика движения точки, где координаты в каждый момент рассчитываются исключительно по формулам движения с ускорением.

Зададим начальные параметры движения: сначала угол α, значение которого должно быть от 1 до 89 включительно:

```python
import math

a = float(input("α =")) / 180 * math.pi

if a <= 0 or a >= math.pi/2:
    a = 1
    print("Ошибка, недопустимое значение. Пожалуйста, введите ещё раз")
```

А также начальную скорость v<span style="font-size:10px;">o</span> - положительное число:

```python
v0 = float(input("V0 ="))
if v0 <= 0:
    v0 = 1
    print("Ошибка, недопустимое значение. Пожалуйста, введите ещё раз")
```

Используя полученные формулы, найдём значение Hmax и Smax

```python
import math

Hmax = v0**2 * math.sin(a)**2 / 20
Smax = v0**2 * math.sin(2 * a) / 10

print("Hmax =", round(Hmax, 3))
print("Smax =", round(Smax, 3))
```

Теперь же посмотрим, как поведёт себя точке при броске с заданными начальными параметрами и посмотрим по графику, будут ли высота и расстояние совпадать

```python
import matplotlib.pyplot as plt

tmax = 2 * v0 * math.sin(a) / 10
tp = [i * tmax / 100 for i in range(101)]

v0x = v0 * math.cos(a)
v0y = v0 * math.sin(a)

x = [v0x * t for t in tp]
y = [v0y * t - 10 * t**2 / 2 for t in tp]

plt.xlabel('x')
plt.ylabel('y')
plt.plot(x, y, label = 'Траектория полёта')
plt.scatter([Smax/2], [Hmax], label="Наивысшая точка", color='red')
plt.scatter([Smax], 0, label="Конечная точка", color = 'green')
plt.legend()
plt.show()
```

Как видно из графика, рассчитанные наивысшая и конечная точка совпадают с фактической траекторией движения

Замечание: при начальных углах α и 90-α дальность полёта будет одинаковой, это связано с тем, что подставив в формулу для расстояния угол α-90, мы получим: sin(2*(90-α)) = sin(180-2α) = sin(2α) по известной формуле приведения

Убедиться в этом можно, построив графики движения при начальных углах α и 90-α при фиксированной начальной скорости на одной плоскости. Введите начальный угол броска α:

```python
a = float(input("α =")) / 180 * math.pi

if a <= 0 or a >= math.pi/2:
    a = 1
    print("Ошибка, недопустимое значение. Пожалуйста, введите ещё раз")
else:
    v0 = 10
    tmax1 = 2 * v0 * math.sin(a) / 10
    tp1 = [i * tmax1 / 100 for i in range(101)]
    
    tmax2 = 2 * v0 * math.sin(math.pi/2 - a) / 10
    tp2 = [i * tmax2 / 100 for i in range(101)]
    
    v0x1 = v0 * math.cos(a)
    v0y1 = v0 * math.sin(a)
    
    v0x2 = v0 * math.cos(math.pi/2 - a)
    v0y2 = v0 * math.sin(math.pi/2 - a)
    
    x1 = [v0x1 * t for t in tp1]
    y1 = [v0y1 * t - 10 * t**2 / 2 for t in tp1]
    
    x2 = [v0x2 * t for t in tp2]
    y2 = [v0y2 * t - 10 * t**2 / 2 for t in tp2]
    
    plt.xlabel('x')
    plt.ylabel('y')
    plt.plot(x1, y1, label = 'Траектория полёта 1')
    plt.plot(x2, y2, label = 'Траектория полёта 2')
    plt.legend()
    plt.show()
```
